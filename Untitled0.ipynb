```python
#data cleaning
#create a data sample
data={
'Name':['Alice','Charlie','David','Edward','Alice'],
'Age':[25,np.nan,45,500,25],
'Salary':['50000','45000','NaN','70000','50000'],
'Department':['HR','IT','HR',None,'HR']
}
df=pd.DataFrame(data)
df
```

```python
#removing duplicates
df.drop_duplicates(inplace=True)
df
```

```python
#handling outlier
df['Age']=np.where(df['Age']>60,df['Age']/10,df['Age'])
df
```

```python
#handling nan values
df['Age'].fillna(df['Age'].mean(),inplace=True)
df
```

```python
#change datatype salary
df['Salary']=df['Salary'].astype(float)
df.info()
```

```python
#handling nan values
df['Salary'].fillna(df['Salary'].mean(),inplace=True)
df
```

```python
#replace none with mode
#handling nan values
df['Department'].fillna(df['Department'].mode()[0],inplace=True)
df
```

```python
df.to_csv('cleaneddata.csv')
```
